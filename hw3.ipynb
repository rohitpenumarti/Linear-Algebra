```python
import numpy as np
from scipy import sparse as sp
from scipy.sparse.linalg import bicg
import matplotlib.pyplot as plt
```

## Part a)

### In this part, I am solving with boundary conditions f(0)=f(pi)=0. I am testing with three n values, 100, 1000, and 10000.

```python
n = 100
y0 = 0
yn = 0

e = np.ones(n)
z = np.zeros(n)
A = sp.spdiags(n**2*np.array([e, -2*e, e])-2*n*np.array([z, -1*e, e]), [-1, 0, 1], n-1, n-1, format='csr')
b = np.cos(np.arange(1, np.pi*n, step=3.17)/n)

y, exitCode = bicg(A, b)

plt.figure(figsize=(5, 5), dpi=150)
plt.plot(np.arange(1, np.pi*n, step=3.17)/n, y, '.')
plt.show()
```

## Part b)

### In this part, I am solving with boundary conditions f(0)= 0 and f(pi)=-1. I am testing with n = 100.

```python
n = 100
y0 = 0
yn = -1

e = np.ones(n)
z = np.zeros(n)
A = sp.spdiags(n**2*np.array([e, -2*e, e])-2*n*np.array([z, -1*e, e]), [-1, 0, 1], n, n, format='csr')
A[98, 99] = 0
A[99, 98] = 0
A[99, 99] = -20400
b = np.cos(np.arange(1, np.pi*n, step=3.14)/n)

y, exitCode = bicg(A, b)

plt.figure(figsize=(5, 5), dpi=150)
plt.plot(np.arange(1, np.pi*n, step=3.14)/n, y, '.')
plt.show()
```
