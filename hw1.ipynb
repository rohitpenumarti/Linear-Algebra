```python
import numpy as np
from time import perf_counter
import matplotlib.pyplot as plt
```

# Problem 5

## Part a)
Below, I created a function to measure the time to compute matrix multiplication between two $n$-dimensional square matrices, where $n$ is specified by the user. Then, I generate an array of $100$ random $n$-values ranging between $100$ and $10000$ and input these values into my function which returns the $n$-value and time taken for computation.

```python
def matrix_multiplication_time(n):
    A = np.random.randn(n, n)
    B = np.random.randn(n, n)

    t1 = perf_counter()
    C = A.dot(B)
    t2 = perf_counter()
    return [n, t2-t1]

n_vals = np.random.randint(100, 10000, 100)
val_pairs = []
for n in n_vals:
    val_pairs.append(matrix_multiplication_time(n))

vals = np.array(val_pairs)
vals
```

Here, I use a linear regression routine to determine the relationship between the log of the $n$-values and the log of the computation times. Below are the results including the estimated values for $c$, $\alpha$, the $n$-value required to multiple two random $nxn$ matrices in one day, and the graph of the relationship between the points.

```python
from scipy.stats import linregress

res = linregress(np.log(vals[:, 0]), np.log(vals[:, 1]))
slope = res.slope
intercept = res.intercept
print('c = {}\n\u03B1 = {}' .format(slope, np.exp(intercept)))

n_1day = int(np.exp((np.log(86400)-intercept)/slope))
print('The n value for which the multiplication between two random nxn matrices will take one day is: ' + str(n_1day))

def line(x):
    return slope*x+intercept

x = np.linspace(np.amin(np.log(vals[:, 0])), np.amax(np.log(vals[:, 0])))
plt.plot(np.log(vals[:, 0]), np.log(vals[:, 1]), 'b.')
plt.plot(x, line(x), 'r-')
plt.xlabel('Log of n-values')
plt.ylabel('Log of time values')
plt.show()
```

## Part b)

## Part c)
Below, I created a function to measure the time to solve a sparse system of $n$ linear equations, where $n$ is specified by the user. I used the spsolve function in the scipy sparse library and also made sure to convert the sparse matrices to csr format. Then, I generate an array of $100$ random $n$-values ranging between $100$ and $10000$ and input these values into my function which returns the $n$-value and time taken for computation.

```python
import scipy.sparse as sp
from scipy.sparse.linalg import spsolve

def sparse_sys_solve(n):
    A = sp.lil_matrix((n, n))
    for i in range(10*n):
        j = np.random.randint(n)
        k = np.random.randint(n)
        A[j, k] = np.random.randn()
    A = A.tocsr()
    b = sp.csr_matrix(np.random.randn(n, 1))

    t1 = perf_counter()
    x = spsolve(A, b)
    t2 = perf_counter()
    return [n, t2-t1]

n_vals = np.random.randint(1, 10000, 100)
val_pairs = []
for n in n_vals:
    val_pairs.append(sparse_sys_solve(n))

vals = np.array(val_pairs)
vals
```

Below are the results including the estimated values for $c$, $\alpha$, the $n$-value required to multiple two random $nxn$ matrices in one day, and the graph of the relationship between the points.

```python
from scipy.stats import linregress

res = linregress(np.log(vals[:, 0]), np.log(vals[:, 1]))
slope = res.slope
intercept = res.intercept
print('c = {}\n\u03B1 = {}' .format(slope, np.exp(intercept)))

n_1day = np.exp((np.log(86400)-intercept)/slope)
print('The n value that would take my computer 24 hours to compute the multiplication between two random nxn matrices is: ' + str(int(n_1day)))

def line(x):
    return slope*x+intercept

x = np.linspace(np.amin(np.log(vals[:, 0])), np.amax(np.log(vals[:, 0])))
plt.plot(np.log(vals[:, 0]), np.log(vals[:, 1]), 'b.')
plt.plot(x, line(x), 'r-')
plt.xlabel('Log of n-values')
plt.ylabel('Log of time values')
plt.show()
```
